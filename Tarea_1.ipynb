```
import numpy as np  
import matplotlib.pyplot as plt
```

```
def step_euler(y,h,f):
  next_y = y+h*f(y)
  return next_y

def euler(h, f, y_0, start_t, end_t):
  N = int((end_t - start_t)/h)
  t_list = np.linspace(start_t, end_t, N+1)
  y_list = np.zeros(N+1)
  y_list[0]=y_0

  for i in range(0,N):
    y_list[i+1] = step_euler(y_list[i], h, f)
  return t_list, y_list

def euler_imp(f,x0,t0,tN,N):
  h=(tN-t0)/N 
  t=np.linspace(t0,tN,N+1)
  x_eu_imp=np.zeros(N+1)
  x_eu_imp[0]=x0
  for n in range(N):
    x_eu_imp[n+1]=x_eu_imp[n]+h*f((x_eu_imp[n]+h*f(x_eu_imp[n],t[n])/2), (t[n]+h/2))
  return t,x_eu_imp

def rk4(f,x0,t0,tN,N):
  t= np.linspace(t0,tN,N+1)
  h= (tN-t0)/N
  x_4rk=np.zeros(N+1)
  x_4rk[0]= x0                         # Condicion inicial

  for n in range(N):
      k1 = h*f( x_4rk[n]       , t[n]         )
      k2 = h*f( x_4rk[n] + k1/2, t[n] + (h/2) ) 
      k3 = h*f( x_4rk[n] + k2/2, t[n] + (h/2) ) 
      k4 = h*f( x_4rk[n] + k3  , t[n] +  h    )
      
      x_4rk[n+1] = x_4rk[n] + k1/6 + k2/3 + k3/3 + k4/6
  return t,x_4rk
```

```
def g(y):
  crecimiento = np.log(2)*y
  return crecimiento

y_0 = 1
h = 0.01
t_0 = 0
t_N = 10

t_list, y_list = euler(h, g, y_0, t_0, t_N)
plt.figure(figsize=(10,6))
plt.plot(t_list, y_list)

plt.plot(t_list, np.power(2, t_list))
plt.grid()
plt.show()
```

## 2.8.3 
(Calibrating the Euler method) The goal of this problem is to test the
Euler method on the initial value problem $\dot{x} = −x$ , $x(0) = 1$.

a) Solve the problem analytically. What is the exact value of $x(1)$?

b) Using the Euler method with step size $\Delta t = 1$, estimate $x(1)$ numerically-call the result $xˆ(1)$ . Then repeat, using $\Delta t = 10^{–n}$, for n  1, 2, 3, 4. 

c) Plot the error $E = |x^(1)-x(1)|$ as a function of $\Delta t$. Then plot $ln( E ) vs. ln (\Delta t)$. Explain the results.

```
def g(y):
  crecimiento = -y
  return crecimiento

X, Y = np.meshgrid(np.linspace(0,6,30), np.linspace(0,1,15))
U = 1
V = g(Y)
fig, ax = plt.subplots(figsize=(20,10))
Q = ax.quiver(X, Y, U, V, width=0.0008)
dt=np.zeros(5)
Err=np.zeros(5)
y_0 = 1
t_0=0
t_N=6
x_1 =np.exp(-1)
for i in range (0,5):
  h=10**-i
  dt[i]=h
  t_list,y_list=euler(h, g, y_0, t_0, t_N)
  ax.plot(t_list,y_list,label=('h=',h))
  xp_1 =y_list[t_list==1]
  E=abs(xp_1-x_1)
  print('Error para h=',h,'=',E)
  Err[i]=E
ax.plot(t_list,np.exp(-t_list),label='Analitico')
ax.axis([t_0,t_N,0,1])
leg = ax.legend()
ax.grid()
plt.figure(figsize=(15,6))
plt.plot(dt[::-1],Err[::-1])
plt.xlabel('Eje de las deltat')
plt.ylabel('Eje de las Error')
plt.grid()
plt.figure(figsize=(15,6))
plt.plot(np.log(dt[::-1]),np.log(Err[::-1]))
plt.xlabel('Eje de las logdeltat')
plt.ylabel('Eje de las logError')
plt.grid()
plt.show()
```
