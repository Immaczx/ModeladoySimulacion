```
import numpy as np  
import matplotlib.pyplot as plt
```

```
def step_euler(y,h,f):
  next_y = y+h*f(y)
  return next_y

def euler(h, f, y_0, start_t, end_t):
  N = int((end_t - start_t)/h)
  t_list = np.linspace(start_t, end_t, N+1)
  y_list = np.zeros(N+1)
  y_list[0]=y_0

  for i in range(0,N):
    y_list[i+1] = step_euler(y_list[i], h, f)
  return t_list, y_list
```

```
def g(y):
  crecimiento = np.log(2)*y
  return crecimiento

y_0 = 1
h = 0.01
t_0 = 0
t_N = 10

t_list, y_list = euler(h, g, y_0, t_0, t_N)
plt.figure(figsize=(10,6))
plt.plot(t_list, y_list)

plt.plot(t_list, np.power(2, t_list))
plt.grid()
plt.show()
```

#Tarea

$$\dot{x}=sin(x)  \quad x_0=0,x_0=\pi/4, x_0=\pi/2$$

```
def g(y):
  crecimiento = np.sin(y)
  return crecimiento

y_ =np.linspace(-2*np.pi,2*np.pi,25)
y_0=[0, np.pi/4, np.pi/2]
h=0.01
t_0=0
t_N=4
plt.figure(figsize=(15,6))
for y in y_ :
  t_list,y_list=euler(h,g,y, t_0,t_N)
  plt.plot(t_list,y_list)
plt.tight_layout()
plt.grid()
```
