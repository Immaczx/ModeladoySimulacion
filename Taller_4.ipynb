```python
import numpy as np
import matplotlib.pyplot as plt
import math
import sympy as sp
from scipy.integrate import solve_ivp
from ipywidgets import interact

def Jacob(fx, fy):
    A = sp.Matrix([[sp.diff(fx, x, 1), sp.diff(fx, y, 1)],
        [sp.diff(fy, x, 1), sp.diff(fy, y, 1)]])
    return A

def Estabilidad(D,T):
  if D > 0 and T > 0:
    return ' Repulsor\n'
  elif D > 0 and T < 0:
    return ' Atractor\n'
  elif D < 0:
    return ' Silla\n'
```

## **6.4.1** $\dot{x}=x(3-x-y)$, $\dot{y}=y(2-x-y)$

```python
from sympy import var, solve
x, y = var('x y')

f1 = -(x**2)+3*x-x*y
f2 = -(y**2)+2*y-x*y

fig, ax = plt.subplots(figsize=(12,8))

sols = solve((f1, f2), (x, y))
print(sols)
print('Se Halla el jacobiano')
A = Jacob(f1, f2)
print(A)
print('Evaluamos los puntos fijos en el jacobiano')
for r in sols:
    print('Se evaluan',r)
    At = A.subs(x, r[0]).subs(y, r[1])
    det = At.det()
    traza = At.trace()
    eigval = At.eigenvals()
    eigvec = At.eigenvects()
    At = np.array([[At[0], At[1]],[At[2], At[3]]])
    print(At)
    print('Eigen valores=',eigval,', con traza =',traza,', con determinante =',det,'y el nodo es:',Estabilidad(det,traza))
    #print('Eigen vectores',eigvec)
    plt.plot(r[0],r[1],'ro')
    plt.annotate(Estabilidad(det,traza),(r[0],r[1]))

X,Y=np.meshgrid(np.linspace(0,4,40),np.linspace(0,3,30))
V= -(X**2)+3*X-X*Y
U= -(Y**2)+2*Y-X*Y
plt.streamplot(X,Y,V, U, color='k', density=1., arrowsize=1)
plt.grid()
plt.plot()
```
