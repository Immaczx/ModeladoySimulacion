##**5.3.3**
En cada uno de los siguientes problemas, predice el curso de la historia de amor, dependiendo de los signos y tamaños relativos de ay b.

(Fuera de contacto con sus propios sentimientos) Supongamos que Romeo y Julieta reaccionan el uno al otro, pero no a sí mismos: $\dot{R}=aJ$ , $\dot{J}= bR$. ¿qué pasa?

```python
for a in [-1,1]:
  for b in [-1,1]:
    A=np.array([[0,a],
              [b,0]])
    delta=np.linalg.det(A)
    traza=np.trace(A)
    print('para a=',a,'y b =',b,'Con traza =',traza,'y con determinante =',delta)
    X,Y=np.meshgrid(np.linspace(-5,5,12),np.linspace(-5,5,12))
    V= b*X
    U= a*Y
    fig, ax = plt.subplots(figsize=(6,6))
    Q= ax.quiver(X,Y,U,V,units ='width')

    def vdp(t,z):
      x,y=z 
      #Donde, j=y
      return[a*y,b*x] #Ecuaciones del sistema

    ti,tf=0,10
    t=np.linspace(ti,tf,500)
    for x0 in range(-5 , 6):
      for y0 in range(-5 , 6):
        for style in zip(styles):
          sol=solve_ivp(vdp,[ti,tf],[x0,y0],t_eval=t)
          plt.plot(sol.y[0],sol.y[1],style)
        plt.axis([-5,5,-5,5])
        plt.grid()
    ax.spines['left'].set_position(('data',0.0))
    ax.spines['bottom'].set_position(('data',0.0))
    ax.spines['top'].set_color('none')
    ax.spines['right'].set_color('none')
    plt.show()
```

##**5.3.5**

(Guisantes en una vaina) Si Romeo y Julieta son clones románticos ($\dot{R} = aR + bJ$, $\dot{J}= bR + aJ$), ¿deberían esperar aburrimiento o felicidad?

```python
valsb = { -2 : [1,2,3], 3 : [1,3,4]}
for a in [-2,3]:
  for b in valsb[a]:
    A=np.array([[a,b],
              [b,a]])
    delta=np.linalg.det(A)
    traza=np.trace(A)
    print('para a=',a,'y b =',b,'Con traza =',traza,'y con determinante =',delta)
    X,Y=np.meshgrid(np.linspace(-5,5,12),np.linspace(-5,5,12))
    V= b*X + a*Y
    U= a*X + b*Y
    fig, ax = plt.subplots(figsize=(6,6))
    Q= ax.quiver(X,Y,U,V,units ='width')

    def vdp(t,z):
      x,y=z 
      #Donde, j=y
      return[a*x+b*y,b*x+a*y] #Ecuaciones del sistema

    ti,tf=0,10
    t=np.linspace(ti,tf,500)
    for x0 in range(-5 , 6):
      for y0 in range(-5 , 6):
        for style in zip(styles):
          sol=solve_ivp(vdp,[ti,tf],[x0,y0],t_eval=t)
          plt.plot(sol.y[0],sol.y[1],style)
        plt.axis([-5,5,-5,5])
        plt.grid()
    ax.spines['left'].set_position(('data',0.0))
    ax.spines['bottom'].set_position(('data',0.0))
    ax.spines['top'].set_color('none')
    ax.spines['right'].set_color('none')
    plt.show()
```
